 2). A company gives a bonus  if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not .

Employee  Sales  Attendance  
Arjun     55000    95
Simran    65000    85
Rohit     30000    90
Anjali    25000    65


            Answer:-  =IF(AND(B3>40000,C3>=90),"Bonus","Not")



3). Write a nested IF formula:  If marks in B2 are ≥ 90 → “A”, ≥ 75 → “B”, ≥ 50 → “C”, otherwise “Fail”.

Student Name  Marks
Raj            85
Neha           96
Vivek          57
Rani           92
Sneha          75



          Answer:-  =IF(B3>90,"A",IF(B3>75,"B",IF(B3>50,"C","Fail")))

EmpID  Name  Department
E101   Raju     IT

* Simple and commonly used for such tables
